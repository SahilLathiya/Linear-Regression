```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300
```

```python
X = np.random.rand(N)*100
```

```python
noise = np.random.uniform(low=0, high=100, size=N)
```

```python
Y = 10*X + 500 + noise
```

```python
w = 0.0
b = 0.0
```

```python
# Define the function to plot
def plot_wx_b(w, b):
  """Plots a line of the form wx + b.

  Args:
    w: The slope of the line.
    b: The y-intercept of the line.
  """

  # Create a range of x-values
  x = np.linspace(0, 100, 100)

  # Calculate the corresponding y-values
  y = w * x + b

  # Plot the line
  plt.plot(x, y)
  # plt.scatter(X[:,1], Y, c='red', edgecolors="black")
  plt.scatter(X, Y, c='red', edgecolors="black")

# Set the x and y labels
plt.xlabel('x')
plt.ylabel('y')
# Call the function to plot the line
plot_wx_b(w, b)

# Show the plot
plt.show()
```

# Stochastic Gradient Descent

```python
epochs = 200
alpha = 0.00025

for i in range(epochs):
    

    for j in range(N):
        Y_pred = w*X[j] + b 
        grad_sum1 = (Y_pred- Y[j])*X[j]
        gradient1 = grad_sum1
        w = w - alpha*gradient1

        grad_sum2 = (Y_pred - Y[j])
        gradient2 = grad_sum2    
        b = b - alpha*gradient2
    
    if i % 10 == 0:
        # Set the x and y labels
        plt.xlabel('x')
        plt.ylabel('y')
        # plt.legend()
        # Call the function to plot the line
        plot_wx_b(w, b)
```

```python
# Set the x and y labels
plt.xlabel('x')
plt.ylabel('y')
# plt.legend()
# Call the function to plot the line
plot_wx_b(w, b)
```

# Batch Gradient Descent

```python
w,b = 0.0, 0.0
```

```python
epochs = 6
alpha = 0.0001

for i in range(epochs):
    
    grad_sum1 = 0
    for j in range(N):
        Y_pred = w*X[j] + b 
        grad_sum1 += (Y_pred- Y[j])*X[j]
    gradient1 = (1/N) * grad_sum1

    grad_sum2 = 0
    for j in range(N):
        Y_pred = w*X[j] + b
        grad_sum2 += (Y_pred - Y[j])
    gradient2 = (1/N) * grad_sum2
    print(w, b)
    
    w = w - alpha*gradient1
    b = b - 0.3*gradient2
    
    if i%10 != 0:
        # Set the x and y labels
        plt.xlabel('x')
        plt.ylabel('y')
        # plt.legend()
        # Call the function to plot the line
        plot_wx_b(w, b)
```

# Using sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data generation
N = 300
X = np.random.rand(N) * 100
noise = np.random.uniform(low=0, high=100, size=N)
Y = 10*X + 500 + noise

# Reshape X for sklearn
X = X.reshape(-1, 1)

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, Y)

# Make predictions
Y_pred = model.predict(X)

# Plot the data and the linear regression line
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()

# Print the coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
