```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 300
```

```python
X = np.random.rand(N)*100
```

```python
noise = np.random.uniform(low=0, high=100, size=N)
```

```python
Y = 10*X + 500 + noise
```

```python
theta = np.array([0.0,0.0])
```

```python
# Define the function to plot
def plot_wx_b(w, b):
  """Plots a line of the form wx + b.

  Args:
    w: The slope of the line.
    b: The y-intercept of the line.
  """

  # Create a range of x-values
  x = np.linspace(0, 100, 100)

  # Calculate the corresponding y-values
  y = w * x + b

  # Plot the line
  plt.plot(x, y)
  # plt.scatter(X[:,1], Y, c='red', edgecolors="black")
  plt.scatter(X, Y, c='red', edgecolors="black")

# Set the x and y labels
plt.xlabel('x')
plt.ylabel('y')
# Call the function to plot the line
plot_wx_b(theta[0], theta[1])

# Show the plot
plt.show()
```

```python
epochs = 5000
alpha = 0.0001

for i in range(epochs):
    
    temp = [theta[0], theta[1]]
    
    grad_sum1 = 0
    for j in range(N):
        grad_sum1 += (temp[1]*X[j] + temp[0] - Y[j])*X[j]
    gradient1 = (1/N) * grad_sum1

    grad_sum2 = 0
    for j in range(N):
        grad_sum2 += (temp[1]*X[j] + temp[0] - Y[j])
    gradient2 = (1/N) * grad_sum2
    
    theta[1] = temp[1] - alpha*gradient1
    theta[0] = temp[0] - alpha*gradient2
```
